```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_books(min_rating=4.0, max_price=20.0):
    # Base URL of the website
    base_url = "http://books.toscrape.com/catalogue/page-{}.html"

    # Empty list to store book data
    books_data = []

    # Create a dictionary to map string ratings to numeric values
    rating_map = {
        'One': 1.0,
        'Two': 2.0,
        'Three': 3.0,
        'Four': 4.0,
        'Five': 5.0
    }

    # Iterate through multiple pages (we'll limit to 5 pages for this example)
    for page in range(1, 6):
        url = base_url.format(page)
        response = requests.get(url)
        
        if response.status_code != 200:
            print(f"Failed to retrieve page {page}")
            continue
        
        # Parse the page with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find all books on the page
        books = soup.find_all('article', class_='product_pod')
        
        # Extract data for each book
        for book in books:
            title = book.find('h3').find('a')['title']
            price = float(book.find('p', class_='price_color').text.replace('Â','').lstrip('£'))  # Clean and convert price
            rating_class = book.find('p', class_='star-rating')['class'][1]  # Rating is stored as class
            rating = rating_map.get(rating_class, 0.0)  # Convert rating to numeric value
            
            # Get the book's link to its detail page
            book_link = book.find('h3').find('a')['href']
            book_detail_url = "http://books.toscrape.com/catalogue/" + book_link.strip('../')  # Build full URL
            
            # Get the genre from the book's detail page
            book_response = requests.get(book_detail_url)
            book_soup = BeautifulSoup(book_response.text, 'html.parser')
            genre = book_soup.find('ul', class_='breadcrumb').find_all('li')[2].text.strip() if book_soup.find('ul', class_='breadcrumb') else 'Unknown'

            # Safe extraction of UPC and Availability
            availability = book.find('p', class_='instock availability').text.strip() if book.find('p', class_='instock availability') else 'N/A'
            upc = 'N/A'  # UPC is not available on this page

            # Filter by the min_rating and max_price
            if rating >= min_rating and price <= max_price:
                books_data.append({
                    'UPC': upc,
                    'Title': title,
                    'Price (£)': price,
                    'Rating': rating,
                    'Genre': genre,
                    'Availability': availability,
                    'Description': 'No description available'  # No description available in this example
                })

    # Create a pandas DataFrame from the list of dictionaries
    df = pd.DataFrame(books_data)

    return df

# Call the function with a rating threshold of 4.0 and max price of 20
books_df = scrape_books(min_rating=4.0, max_price=20.0)

# Display the first 5 rows of the DataFrame
display(books_df.head())
```
